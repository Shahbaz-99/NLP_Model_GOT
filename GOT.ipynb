```python
import os
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess
```

Preprocessing text on GOT books

```python


def load_and_preprocess_corpus(dataset_path: str) -> list[list[str]]:
    stop_words = set(stopwords.words('english'))  # all lowercase
    story = []

    for filename in os.listdir(dataset_path):
        filepath = os.path.join(dataset_path, filename)

        with open(filepath, encoding="utf-8", errors="ignore") as f:
            corpus = f.read()

        for sent in sent_tokenize(corpus):
            tokens = simple_preprocess(sent)  # lowercases + tokenizes first
            filtered_tokens = [word for word in tokens if word not in stop_words]  # then filter
            
            if filtered_tokens:  # skip empty sentences
                story.append(filtered_tokens)

    return story

story = load_and_preprocess_corpus(r"C:\Official_stuffs\AI-ML\datasets\game-of-thrones-books")
# print(story[:500])
```

```python
model = gensim.models.Word2Vec(   
    window = 8 ,  # how many context words
    min_count = 10,  # we cosider that sentence which have at least 2 words

)
```

```python
model.build_vocab(story)
```

```python
model.train(story,
            total_examples = model.corpus_count,
            epochs = 5
            )
```
